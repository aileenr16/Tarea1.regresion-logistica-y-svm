```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
import seaborn as sns
from sklearn import metrics
from sklearn.svm import LinearSVC
from sklearn.model_selection import cross_val_score
import statistics

import warnings
from sklearn.datasets import fetch_openml
from sklearn.exceptions import ConvergenceWarning
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
```

```python
data = pd.read_csv('C:\\Users\\ailee\Desktop\\INTRO_CIENC_DATOS\\Clase_3\\week2.csv')
data.reset_index(inplace=True)
data.columns = ['X1', 'X2', 'y']
data.head()
```

```python
#se hace una copia para no modificar la data original 
df = data.copy()
df.head()
```

```python
X1=df.iloc[:,0]
X2=df.iloc[:,1]
X=np.column_stack((X1,X2))
y=df.iloc[:,2]

plt.scatter(X1[y==1], X2[y==1], c='g', marker = '+', label='1')
plt.scatter(X1[y==-1], X2[y==-1], c='b', marker = 'o', label='-1', s=10)
plt.xlabel('x_1')
plt.ylabel('x_2')
plt.legend(bbox_to_anchor=(1.15,1.15), loc='upper right', fancybox=True, framealpha=1, fontsize=10)
# plt.show()
plt.title('Visualización de los datos') 
plt.savefig('Figure_1.png')

plt.show()
```

```python
#división de los conjuntos
#se trabaja con un 80% datos de entrenamiento y 20% datos de prueba

x_train_1, x_test_1, y_train_1, y_test_1 = train_test_split(X,y, test_size=0.2,random_state=4)
print('Train Set: ', x_train_1.shape, y_train_1.shape)
print('Test Set: ', x_test_1.shape, y_test_1.shape)
```

```python
#entrenamiento del modelo

LR_1=LogisticRegression()
LR_1.fit(x_train_1, y_train_1)
print('The slopes are: ',LR_1.coef_[0])
print('The intercept is: ',LR_1.intercept_)
predictions_1 = LR_1.predict(x_train_1)
score_1 = LR_1.score(x_train_1, y_train_1)
print('The score is: ',score_1)
```

```python
#metricas del conjunto y graficas

feature_importance_1 = LR_1.coef_[0]
for i,val in enumerate(feature_importance_1):
    print('Feature: %0d, Score: %.5f' % (i,val))
plt.bar([x for x in range(len(feature_importance_1))], feature_importance_1)
plt.xlabel('Features')
plt.ylabel('Scores')
# plt.show()
#plt.savefig('Figure_2.png')

cm_1 = metrics.confusion_matrix(y_train_1, predictions_1)
print(cm_1)

# Report
cr_1 = metrics.classification_report(y_train_1, predictions_1)
print(cr_1)


plt.figure(figsize=(6,6))
sns.heatmap(cm_1, annot=True, fmt=".0f", linewidths=.5, square = True, cmap = 'Greens');
plt.ylabel('Actual Value');
plt.xlabel('Predicted Value');
all_sample_title_1 = 'Accuracy: {0}'.format(score_1)
plt.title(all_sample_title_1, size = 15);
plt.savefig('Figure_4.png')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.svm import LinearSVC
from sklearn import metrics

# Define la lista de valores de C
C_values = [0.001, 0.1, 1, 10, 100, 1000]

# Itera sobre cada valor de C
for C in C_values:
    vector_m = LinearSVC(C=C)
    model_vector_m = vector_m.fit(x_train_1, y_train_1)
    print(f"For C = {C}:")
    print("Los coeficientes son:", model_vector_m.coef_[0])
    print("El intercepto es:", model_vector_m.intercept_)
    
    predictions_vm = model_vector_m.predict(x_train_1)
    score_vm = model_vector_m.score(x_train_1, y_train_1)
    print("El puntaje es:", score_vm)
    print("\n")

    # Calcular sesgo y varianza
    y_train_mean = np.mean(y_train_1)
    bias = np.mean(predictions_vm) - y_train_mean
    variance = np.var(predictions_vm)

    # Pérdida esperada
    expected_loss = bias**2 + variance
    
    print(f"Sesgo: {bias:.4f}")
    print(f"Varianza: {variance:.4f}")
    print(f"Pérdida esperada: {expected_loss:.4f}")
    print("\n" + "="*50 + "\n")

 
```
